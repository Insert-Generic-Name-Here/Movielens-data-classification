```python
import pickle

from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
```

```python
ratings = open('ml-100k-temporary-data/ratings-np-array.pkl','rb')
ratings = pickle.load(ratings)
```

```python
print (ratings)
```

```python
# clustering dataset
# determine k using elbow method
def estimate_clusters(X, kMax, width, height):
    inertia = []
    kTest = range(1, kMax)
    
    for i in kTest:
        tmp = KMeans(n_clusters=i, init='k-means++').fit(X)
        inertia.append(tmp.inertia_)
        
    plt.figure(figsize=(width, height))
    plt.plot(kTest, inertia, 'bx-')    
    plt.xlabel('k')
    plt.ylabel('Inertia')
    plt.title('The Elbow Method showing the optimal k')
    plt.grid()
    plt.show()    
```

```python
estimate_clusters(ratings, 50, 50, 50)
# 11 Clusters?????
```
